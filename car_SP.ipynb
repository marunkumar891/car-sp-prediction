```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
car_data = pd.read_csv('car data.csv')
```

```python
car_data.head()
```

```python
car_data = car_data.drop(['Car_Name'],axis=1)
car_data.head()
```

<h3>Selling_price attribute is the dependant variable</h3>

```python
print(car_data['Fuel_Type'].unique())
print(car_data['Seller_Type'].unique())
print(car_data['Transmission'].unique())
print(car_data['Owner'].unique())
```

```python
car_data['cur_year'] = 2020
car_data.head()
```

```python
car_data['Age'] = car_data['cur_year']-car_data['Year']
car_data.head(10)
```

```python
car_data = car_data.drop(['Year', 'cur_year'], axis = 1)
car_data.tail()
```

```python
car_data.describe()
```

```python
car_data.isna().any()
```

```python
car_data.info()
```

```python
print(car_data['Fuel_Type'].value_counts(), '\n')
print(car_data['Seller_Type'].value_counts(), '\n')
print(car_data['Transmission'].value_counts(), '\n')
print(car_data['Owner'].value_counts())
```

```python
car_data = pd.get_dummies(car_data,drop_first=True)
car_data.head(10)
```

```python
car_data.corr()
```

```python
sns.pairplot(car_data)
```

```python
matrix = car_data.corr(method = 'pearson')
fig = plt.figure(figsize = (10,10))

sns.heatmap(matrix, square = True, vmax = 0.8, annot = True)
plt.show()
```

```python
car_data.hist(bins=50, figsize= (15,15))
plt.show()
```

```python
y = car_data.iloc[:,0]
X = car_data.iloc[:,1:]

print(X.shape)
print(y.shape)

print("Independant varibales : \n",X.head())
print("Target variable : \n",y.head())
```

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python

x_train,x_test,y_train,y_test=tts(X,y,test_size=0.2,random_state=1)
print("x train: ",x_train.shape)
print("x test: ",x_test.shape)
print("y train: ",y_train.shape)
print("y test: ",y_test.shape)
```

```python
lr.fit(x_train,y_train)

predicts=lr.predict(x_test)

prediction=pd.DataFrame(predicts)

R_2=r2_score(y_test,prediction)

print("Linear Regression Model :") 
print("r_2 score :",R_2,"\n")
```

```python
sns.distplot(y_test-prediction)
plt.show()
plt.scatter(y_test,prediction)
plt.show()
```
